**Goal of this notebook?**
The goal of this notebook is to help a bank marketing campaign to have a better successful rate for attracting more clients to subscribe a term deposit. Lets quicky explain how subscribing a deposit works how bank and client benefit from it? 

**A term deposit is a type of deposit account held at a financial institution where money is locked up for some set period of time, it could range months to years.**

**How clientâ€™s benefits?**
Typically, term deposits offer higher interest rates than traditional liquid savings accounts, whereby customers can withdraw their money at any time.

**why clients might not like ?**
If a customer wants to close a term deposit before the end of the term, or maturity, the customer will be subject to a penalty. This penalty may include the loss of any interest paid on the deposit account until that point.

**What population of clients more likely to subscribe?**
Term deposits are an extremely safe investment and are therefore very appealing to **conservative, low-risk investors**.

**How will bank benefit?**
If a customer places money in a term deposit,**the bank can invest the money in other financial products that pay a higher rate of return (RoR) than what the bank is paying the customer for the use of their funds. The bank can also lend the money out to its other clients**, thereby receiving a higher interest rate from the borrowers as compared to what the bank is paying in interest for the term deposit.

To see if there is any relation between some multi-category varaible like job and month with target! I am going to simply find success deposit in that group and divide by total number of phone call in that group! and then compares two other groups 
 for example, for month of Aril, I count number of success phone calls in April divided by total number of calls made in April and repeat for othet months and compar them together! if success rate was more or less the same for different ,onths. then probably months couldnt be a very important feature

```python
list_month=df['contact month'].unique().tolist()
success_month=[]
attemp_rate=[]
for month in list_month:

    success_rate=100*df.groupby(['contact month','y']).size()[month]['yes']/df.groupby(['contact month','y']).size()[month].sum()
    
    total_Attempt=100*df['contact month'].value_counts()[month]/df['contact month'].value_counts().sum()
    
    success_month.append(success_rate)
    attemp_rate.append(total_Attempt)
```

```python

fig = plt.figure(figsize=(10,3))
ax = fig.add_axes([0,0,1,1])
ax.bar(list_month,success_month)
plt.ylabel('succes contact rate per month %')
plt.xticks(rotation=90)
plt.show()
```

# Baseline Model

**cross validation is used to avoid overfitting**

```python
LogR=LogisticRegression(solver="liblinear",random_state=42)
accuracy = cross_val_score(LogR, X_train, y_train, cv=3, scoring='accuracy')
Mean=100*accuracy.mean()
SD =100*accuracy.std()
print( "  Mean Accuracy: ", Mean, "   Standard Deviation: ", SD)
```

**what evaluation metrics should be used and why?**
accuracy is high but here accuracy is not a really good evaluation metric, because there are few people subscribed the term deposit and we have an unbalance data!
 Accuracy shows percentage of correct prediction for people who made prediction and people who didnt .But here we have around 90% of people who didnt make any deposit so if we had a extremely simple model that predicted all outcome to be yes  then still accuracy was 90%!

 Instead we  should use other metrics like  recall and precision. Recall show how many of people who actually make a deposit had been detected in the model and precision shows how many of people labeled as â€˜made a depositâ€™ in the model, actually  made any deposit in real world! and f1_score is a combination of these two metrics if both recall and precision were high, f1_score will be high too!

```python

y_pred = cross_val_predict(LogR, X_train, y_train, cv=3)
F1_score=f1_score(y_train, y_pred)
print( 'F1_score:',100*F1_score )
```

```python
ConM=confusion_matrix(y_train, y_pred)
sns.heatmap(ConM,annot=True)
```
